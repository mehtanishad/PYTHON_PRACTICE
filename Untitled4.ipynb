```python
class Banker:
    def __init__(self, name, account_number, balance):
        self.name = name
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Amount {amount} deposited successfully. Current balance is {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print(f"Amount {amount} withdrawn successfully. Current balance is {self.balance}")

    def display_balance(self):
        print(f"Current balance for account {self.account_number} is {self.balance}")


class Customer:
    def __init__(self, name, account_number):
        self.name = name
        self.account_number = account_number

    def deposit(self, banker, amount):
        banker.deposit(amount)

    def withdraw(self, banker, amount):
        banker.withdraw(amount)

    def check_balance(self, banker):
        banker.display_balance()


def main():
    banker = Banker("John Doe", "1234567890", 1000)
    customer = Customer("Jane Doe", "0987654321")

    while True:
        print("Welcome to the bank management console application!")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check balance")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            amount = int(input("Enter the deposit amount: "))
            customer.deposit(banker, amount)

        elif choice == "2":
            amount = int(input("Enter the withdrawal amount: "))
            customer.withdraw(banker, amount)

        elif choice == "3":
            customer.check_balance(banker)

        elif choice == "4":
            print("Thank you for using the bank management console application!")
            break

        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
